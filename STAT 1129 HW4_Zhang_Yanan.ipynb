```python
#3. using x integer arrary as X-axis,using y integer arrary as Y-axis, plot a line.
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([1,6])
y = np.array([5,10])

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show
```

```python
#4.Plot a figure with multiple lines, as shown in figure 2.
import matplotlib.pyplot as plt
import numpy as np

y1=np.array([2,7])
y2=np.array([3,8])
y3=np.array([4,9])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.xlabel("X Label")
plt.ylabel("Y Label")

plt.show
```

```python
#5.practice subplots()function, design your own subplot.
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([3,6,9,12])
y = np.array([1,3,5,7])

plt.subplot(1,3,1)
plt.plot(x,y)
plt.title("Profit")

#plot 2:
x = np.array([2,4,8,10])
y = np.array([20,30,40,50])

plt.subplot(1,3,2)
plt.plot(x,y)
plt.title("Loss")

plt.show()
```
