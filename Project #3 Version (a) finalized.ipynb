```python
import random #import random for computer to generate its choice

class RockPaperScissors:
    #Class to deal with an instance of a Rock-Paper-Scissors game with unlimited rounds.

    def __init__(self): #Initialize the variables for the class and set up welcome words. 
        self.ties = 0
        self.losses = 0
        self.wins = 0
        print("Welcome to Rock, Paper, Scissors Game!\n")
    
    def usermove(self):
        user_choice = input("Choose: Rock, Paper, or Scissors") #lets user choose what move to play
        if user_choice == "Rock":
            user_move = "Rock" #checks if the user chose rock, and assigns it
            print("You chose",user_move)
            return user_move
        elif user_choice == "Paper":
            user_move = "Paper"
            print("You chose",user_move)
            return user_move
        elif user_choice == "Scissors":
            user_move = "Scissors"
            print("You chose",user_move)
            return user_move
        else:
            user_move = "Error" #check correctness of input, and lets user know they did the input wrong
            print("Not a valid option. Did you remember to capitalize?")
            return user_move

    def randomcomp(self):
        moves = ["Rock", "Paper", "Scissors"]
        comp_move = random.choice(moves) #randomly chooses one of the 3 moves
        return comp_move

    def results(self, user, computer):
        print("Computer chose", computer)
        if user == "Rock": #Outputs the result if user chose rock
            if computer == "Rock":
                print("It's a tie!")
                self.ties += 1
            elif computer == "Paper":
                print("Sorry,You lose!")
                self.losses += 1
            elif computer == "Scissors":
                print("Congratulation, You win!")
                self.wins += 1
            else:
                print("There was an error, and the game could not be played.")
        elif user == "Paper":
            if computer == "Rock":
                print("Congratulation,You win!")
                self.wins += 1
            elif computer == "Paper":
                print("It's a tie!")
                self.ties += 1
            elif computer == "Scissors":
                print("Sorry,You lose!")
                self.losses += 1
            else:
                print("There was an error, and the game could not be played.")
        elif user == "Scissors":
            if computer == "Rock":
                print("Sorry,You lose!")
                self.losses += 1
            elif computer == "Paper":
                print("Congratulation,You win!")
                self.wins += 1
            elif computer == "Scissors":
                print("It's a tie!")
                self.ties += 1
            else:
                print("There was an error, and the game could not be played.")
        else: 
            print ("There was an error, and the game could not be played.") #outputs if there was something wrong with user's input
        
    def scoreboard(self):
        print ("You have", self.wins, "wins,", self.losses, "losses,", "and", self.ties, "ties")   
        
if __name__ == "__main__":
    # Initialize an instance of RockPaperScissors for user to refer to
    game = RockPaperScissors() # Keep playing the came repeatedly if user chooses"yes", stop playing if user chooses "no"

    while True: # Find out if the user want to continue playing or not.
        user = game.usermove()   
        computer = game.randomcomp()
        game.results(user, computer)
        game.scoreboard()
        while True:
            continue_display = input('\nDo you want to play again? (yes/no): ').lower()
            if continue_display == 'no':
                # Stop the game after printing a response.
                print("Bye, see you next time!")
            elif continue_display == 'yes':
                # Break the continue prompt loop and keep playing.
                break
            else:
                # invalid input was given, request if user wants to play again.
                print("Invalid input!\n")
                continue
```
