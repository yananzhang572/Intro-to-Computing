```python
import random


class RockPaperScissors:
    #Class to handle an instance of a Rock-Paper-Scissors game with unlimited rounds.

    def __init__(self): #Initialize the variables for the class
        self.ties = 0
        self.losses = 0
        self.wins = 0
        self.options = {'rock': 1, 'paper': 2, 'scissors': 3}

    def random_choice(self): #Chooses a choice randomly from the keys in self.options
   
        return random.choice(list(self.options.keys())) 
    #returns: String containing the choice of the computer.

    def calculate_win(self, user, computer): #Check if the user wins or loses.
        #param user: Numeric representation of player choice from self.options
        #param computer: Numeric representation of computer choice from self.options
        #return: Nothing, but will print whether win or lose
        score = (user - computer) % 3
        if score == 1:
            self.ties += 1
            print("The game is a tie!")
        elif score == 2:
            self.wins += 1
            print("Congraulation, You win!")
        elif score == 3:
            self.losses += 1
            print("Sorry, You lose!")

    def print_score(self): #Prints a string reflecting the current player score.
        #return: Nothing, just prints current score.
        print(f"You have {self.ties} ties, {self.losses} losses, and "
              f"{self.wins} wins.")

    def play_game(self): #Plays a round of Rock-Paper-Scissors with the computer.
        #return: Nothing
        while True:
            userchoice = input("Choices:'rock', 'paper', or 'scissors'.\n"
                               "Which do you choose? ").lower()
            if userchoice not in self.options.keys():
                print("Error, please try again!")
            else:
                break
        computer_choice = self.random_choice()
        print(f"You've picked {userchoice}, and I picked {computer_choice}.")
        self.calculate_win(self.options[userchoice], self.options[computer_choice])


if __name__ == "__main__": # Initialize an instance of RockPaperScissors for people to refer to
    game = RockPaperScissors() # Keep playing the came repeatedly, stop playing by exiting
    # the entire program directly.
    while True:
        game.play_game()  # Run a round of RPS
        game.print_score()  # Print the score(s) after the round

        # Find out if the user want to continue playing or not.
        while True:
            continue_prompt = input('\nDo you wish to play again? (y/n): ').lower()
            if continue_prompt == 'no':
                # Exit the game directly after printing a response.
                print("Bye, see you next time!")
                exit()
            elif continue_prompt == 'yes':
                # Break the continue prompt loop and keep playing.
                break
            else:
                # Bad input was given, re-request if we want to play again.
                print("Invalid input!\n")
                continue
```
